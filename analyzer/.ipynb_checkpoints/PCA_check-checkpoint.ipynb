```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
# pickleからデータを開く
with open('kampo_list.pickle', mode='rb') as f:
    keme = pickle.load(f)
# group.name, user_count, start_day, user.username, steps, distance, kampo, step1, step2, ... , step10
```

```python
mikan_data = keme[np.where( keme[:, 6] == False ), 7:17]
kampo_data = keme[np.where( keme[:, 6] == True ), 7:17]
mikan_data = mikan_data[0, ]
kampo_data = kampo_data[0, ]
```

# 色々結果が怪しいので主成分分析してみる

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(mikan_data)
plt.xlim(-1.0, 1.0)
plt.ylim(-1.0, 1.0)
plt.scatter(pca.components_[0], pca.components_[1])
plt.show()
```

```python
pca = PCA(n_components=4)
pca.fit(kampo_data)
plt.xlim(-1.0, 1.0)
plt.ylim(-1.0, 1.0)
plt.scatter(pca.components_[0], pca.components_[1])
plt.show()
plt.xlim(-1.0, 1.0)
plt.ylim(-1.0, 1.0)
plt.scatter(pca.components_[1], pca.components_[2])
plt.show()
plt.xlim(-1.0, 1.0)
plt.ylim(-1.0, 1.0)
plt.scatter(pca.components_[2], pca.components_[3])
plt.show()
# plt.scatter(pca.components_[3], pca.components_[4])
# plt.show()
print len(kampo_data), pca.explained_variance_ratio_
```

```python
pca  = PCA(1)
result = pca.fit_transform(kampo_data)
plt.scatter(range(len(result)), result)
print len(kampo_data), pca.explained_variance_ratio_
```

```python
pca  = PCA(1)
result = pca.fit_transform(mikan_data)
plt.scatter(range(len(result)), result)
```

# DBscanでクラスタリング

```python
db = DBSCAN(eps=0.01).fit(kampo_data)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
```

```python
db = DBSCAN(eps=0.01).fit(kampo_data)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
plt.xlim(-5000, 50000)
plt.ylim(-5000, 50000)
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)

    xy = kampo_data[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14)

    xy = kampo_data[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```
