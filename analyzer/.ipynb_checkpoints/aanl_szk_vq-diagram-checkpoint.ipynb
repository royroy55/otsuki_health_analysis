## 状態遷移図を描く関数

```python
import networkx as nx
import numpy
import scipy.cluster
from sklearn.preprocessing import normalize

def combochart(P, step_codebook, original_code, nrank=4, thresh=None, max_width=4.):
    G = nx.DiGraph()
    vec = numpy.copy(P)
    vec.shape = vec.shape[0] * vec.shape[1],
    N = nrank + 1 # plus noise class
    codebook, destortion = scipy.cluster.vq.kmeans(vec, N, iter=20)
    codebook = sorted(codebook)
    quantities = normalize(
            [numpy.count_nonzero((original_code == i)) for i in range(len(step_codebook))]
    )[0] * max_width
    for i in numpy.arange(P.shape[0]):
        node_label = '%d: %d' % (i, step_codebook[i])
        print 'quantity:', quantities[i]
        G.add_node(i, label=node_label, shape='circle', width=quantities[i])
        for i in xrange(P.shape[0]):
            ranks, dist = scipy.cluster.vq.vq(P[i], codebook)
            for j in xrange(P.shape[1]):
                rank = ranks[j]
                if rank == 0: continue
                if not thresh is None and P[i][j] < thresh: continue
                label = ' %.2f' % P[i][j] if P[i][j] > .01 else ' %.2e' % P[i][j]
                G.add_edge(
                    i, j,
                    penwidth=rank,
                    label=label,
                    color='black' if i == j else 'blue' if i < j else 'red'
                )
    return G
```
